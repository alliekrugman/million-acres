```python
#import pandas
import pandas as pd
```

```python
#import csv
df_aqi = pd.read_csv('daily_aqi_by_county_2020.csv')
```

```python
#create dataframe
df_aqi
```

```python
#isolate Mendocino county
mendocino = df_aqi.loc[df_aqi['county Name'] == 'Mendocino']
```

```python
#show new dataframe
mendocino
```

```python
#save new dataframe
mendocino.to_csv("mendocino_aqi.csv", index=False)
```

```python
mendocino.loc[:, 'Rolling AQI'] = mendocino['AQI'].rolling(window=10).mean()
mendocino = mendocino.dropna()
```

```python
mendocino.to_csv("mendocino_aqi.csv", index=False)
```
